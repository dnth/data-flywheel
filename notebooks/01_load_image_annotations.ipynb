```python
bounding_box_annotator = sv.BoundingBoxAnnotator()
label_annotator = sv.LabelAnnotator()
```

```python
import cv2
image = cv2.imread('/workspace/datasets/coco_minitrain_25k/images/train2017/000000231538.jpg')

detections = dataset.annotations['/workspace/datasets/coco_minitrain_25k/images/train2017/000000231538.jpg']

```

```python
len(detections)
```

```python
detections.class_id
```

```python
labels = [dataset.classes[id] for id in detections.class_id]
labels
```

```python
annotated_image = bounding_box_annotator.annotate(scene=image, detections=detections)
annotated_image = label_annotator.annotate(scene=image, detections=detections, labels=labels)
```

```python
sv.plot_image(annotated_image)
```
